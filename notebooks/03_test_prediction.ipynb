```python
import pandas as pd
import joblib
```

```python
model = joblib.load("../models/food_drug_risk_model.pkl")
encoders = joblib.load("../models/encoders.pkl")
```

```python
features = [
    "Chemical_Class",
    "Habit_Forming",
    "Therapeutic_Class",
    "Action_Class",
    "energy",
    "protein",
    "fat",
    "carbs",
    "fiber",
    "calcium",
    "iron",
    "vitamin_c",
    "vitamin_a",
    "vitamin_k_proxy",
    "is_alcohol",
    "is_leafy_green",
    "meal_type"
]
```

```python
sample = {
    "Chemical_Class": "716",
    "Habit_Forming": 1,
    "Therapeutic_Class": "11",
    "Action_Class": "190",
    "energy": 350,
    "protein": 8,
    "fat": 20,
    "carbs": 60,
    "fiber": 3,
    "calcium": 20,
    "iron": 2,
    "vitamin_c": 5,
    "vitamin_a": 10,
    "vitamin_k_proxy": 0,
    "is_alcohol": 0,
    "is_leafy_green": 0,
    "meal_type": "Dinner"
}
```

```python
sample_df = pd.DataFrame([sample])
```

```python
for col, le in encoders.items():
    sample_df[col] = sample_df[col].astype(str)
    known_classes = set(le.classes_)
    sample_df[col] = sample_df[col].apply(lambda x: x if x in known_classes else le.classes_[0])
    sample_df[col] = le.transform(sample_df[col])

prediction = int(model.predict(sample_df[features])[0])
probability = model.predict_proba(sample_df[features])[0]

label_map = {
    0: "Safe",
    1: "Moderate",
    2: "Dangerous"
}
```

```python
print("Prediction:", label_map[prediction])
print("Probability:", probability)
```
