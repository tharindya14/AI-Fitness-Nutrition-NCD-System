```python
import pandas as pd
```

```python
import os
```

```python
import pickle
```

```python
DATASET_PATH = "../datasets/supplement_dataset"
MODEL_PATH = "../models/supplement"

model_path = os.path.join(MODEL_PATH, "supplement_model.pkl")
supplements_path = os.path.join(DATASET_PATH, "supplements_sri_lankan_prices.csv")
interactions_path = os.path.join(DATASET_PATH, "interactions.csv")
```

```python
with open(model_path, "rb") as f:
    model = pickle.load(f)

supplements_df = pd.read_csv(supplements_path)
interactions_df = pd.read_csv(interactions_path)

print("Model and datasets loaded successfully!")
print("Supplements:", supplements_df.shape)
print("Interactions:", interactions_df.shape)
```

```python
sample_user = pd.DataFrame([{
    "age": 24,
    "gender": "female",
    "height_cm": 160,
    "weight_kg": 55,
    "bmi": 21.5,
    "health_condition": "none",
    "medication": "none",
    "goal": "muscle gain",
    "budget_range": "10000-25000",
    "activity_level": "moderate"
}])
```

```python
predicted_category = model.predict(sample_user)[0]

print("Predicted Category:", predicted_category)
```

```python
supplements_df.columns.tolist()
```

```python
# Normalize text columns
text_cols = ["category", "price_range_lkr", "ingredients", "brand", "name", "target_goal"]

for col in text_cols:
    if col in supplements_df.columns:
        supplements_df[col] = supplements_df[col].astype(str).str.strip().str.lower()

predicted_category = str(predicted_category).strip().lower()
user_budget = str(sample_user.loc[0, "budget_range"]).strip().lower()

print("Predicted Category:", predicted_category)
print("User Budget:", user_budget)
```

```python
recommended_products = supplements_df[
    (supplements_df["category"] == predicted_category) &
    (supplements_df["price_range_lkr"] == user_budget)
]

print("Recommended products count:", len(recommended_products))

recommended_products.head(10)
```

```python
# Normalize interaction dataset text columns
for col in interactions_df.columns:
    if interactions_df[col].dtype == "object":
        interactions_df[col] = interactions_df[col].astype(str).str.strip().str.lower()

interactions_df.head()
```

```python
def check_supplement_safety(recommended_products, user_medications, interactions_df):
    safe_products = []
    unsafe_products = []

    # user medications normalize
    user_medications = [str(med).strip().lower() for med in user_medications]

    for _, product in recommended_products.iterrows():
        product_ingredients = str(product.get("ingredients", "")).lower()
        product_name = product.get("name", "unknown product")

        product_warnings = []

        for med in user_medications:
            if med in ["none", "no", "nan", ""]:
                continue

            # Find interaction rules for this medication
            med_rules = interactions_df[
                interactions_df["medicine"].str.contains(med, case=False, na=False)
            ]

            for _, rule in med_rules.iterrows():
                risky_ingredient = str(rule["supplement_ingredient"]).lower()

                # Check if risky ingredient exists in product ingredients
                if risky_ingredient in product_ingredients:
                    product_warnings.append({
                        "medicine": rule["medicine"],
                        "supplement_ingredient": rule["supplement_ingredient"],
                        "interaction_risk": rule["interaction_risk"],
                        "severity": rule["severity"],
                        "recommendation": rule["recommendation"]
                    })

        if product_warnings:
            unsafe_products.append({
                "supplement_id": product.get("supplement_id"),
                "name": product_name,
                "category": product.get("category"),
                "price_lkr": product.get("price_lkr"),
                "warnings": product_warnings
            })
        else:
            safe_products.append(product)

    safe_products_df = pd.DataFrame(safe_products)

    return safe_products_df, unsafe_products
```

```python
user_medications = ["none"]

safe_products_df, unsafe_products = check_supplement_safety(
    recommended_products,
    user_medications,
    interactions_df
)

print("Safe products:", len(safe_products_df))
print("Unsafe products:", len(unsafe_products))

safe_products_df.head(5)
```
