```python
import pandas as pd
import os
```

```python
DATASET_PATH = "../datasets/supplement_dataset"

supplements_path = os.path.join(DATASET_PATH, "supplements_sri_lankan_prices.csv")
user_profiles_path = os.path.join(DATASET_PATH, "user_profiles.csv")
interactions_path = os.path.join(DATASET_PATH, "interactions.csv")
schedule_rules_path = os.path.join(DATASET_PATH, "schedule_rules.csv")
progress_logs_path = os.path.join(DATASET_PATH, "progress_logs.csv")
```

```python
supplements_df = pd.read_csv(supplements_path)
user_profiles_df = pd.read_csv(user_profiles_path)
interactions_df = pd.read_csv(interactions_path)
schedule_rules_df = pd.read_csv(schedule_rules_path)
progress_logs_df = pd.read_csv(progress_logs_path)

print("All datasets loaded successfully!")
```
